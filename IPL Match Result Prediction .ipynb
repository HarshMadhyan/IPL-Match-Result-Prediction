*Train-Test Split*

```python
from sklearn.model_selection import train_test_split
y = df['Winner (team 1=1, team 2=0)']
```

```python
X=df.drop(['Winner (team 1=1, team 2=0)'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.metrics import accuracy_score, log_loss, mean_squared_error as mse
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
import lightgbm as lgb
from xgboost import XGBClassifier
```

*Applying and testing models*

```python
import math
classifiers = [DecisionTreeClassifier(), RandomForestClassifier(), AdaBoostClassifier(), GradientBoostingClassifier(), LogisticRegression(penalty='l2'), LogisticRegression(penalty='l1')]
for clf in classifiers:
    clf.fit(X_train, y_train)
    print(clf)
    y_pred = clf.predict(X_test)
    print('Accuracy:', accuracy_score(y_pred, y_test))
    print('Log_loss:', log_loss(y_pred, y_test))
    print('RMSE:', math.sqrt(mse(y_pred, y_test)))
```

```python
clf1 = lgb.LGBMClassifier()
clf1.fit(X_train, y_train)
y_pred1 = clf1.predict(X_test)
print('Accuracy:', accuracy_score(y_pred1, y_test))
print('Log_loss:', log_loss(y_pred1, y_test))
print('RMSE:', math.sqrt(mse(y_pred1, y_test)))
```

```python
from sklearn.neural_network import MLPClassifier
mlp=MLPClassifier()
mlp.fit(X_train, y_train)
y_pred2 = mlp.predict(X_test)
print('Accuracy:', accuracy_score(y_pred2, y_test))
print('RMSE:', math.sqrt(mse(y_pred2, y_test)))
```

*Trying with Ensembles*

```python
from sklearn.ensemble import VotingClassifier
estimators=[('LGBM', lgb.LGBMClassifier()), ('LOGReg', LogisticRegression(penalty='l2')), ('AdaBoost', AdaBoostClassifier())]
ensemble1 = VotingClassifier(estimators, voting='hard')
ensemble1.fit(X_train, y_train)
ensemble1.score(X_test, y_test)
```

```python
from mlens.ensemble import SuperLearner
ensemble2 = SuperLearner()
ensemble2.add([lgb.LGBMClassifier(), LogisticRegression(penalty='l2'), AdaBoostClassifier(), GradientBoostingClassifier()])
ensemble2.add_meta(lgb.LGBMClassifier())
ensemble2.fit(X_train, y_train)
preds = ensemble2.predict(X_test)
math.sqrt(mse(y_test, preds))
accuracy_score(y_test, preds)
```

```python
ensemble3 = SuperLearner()
ensemble3.add([lgb.LGBMClassifier(), LogisticRegression(penalty='l2'), lgb.LGBMClassifier(), LogisticRegression(penalty='l2')])
ensemble3.add_meta(lgb.LGBMClassifier())
ensemble3.fit(X_train, y_train)
preds1 = ensemble3.predict(X_test)
math.sqrt(mse(y_test, preds1))
accuracy_score(y_test, preds1)
```

```python
estimators4=[('LGBM', lgb.LGBMClassifier()), ('LOGReg', LogisticRegression(penalty='l1'))]
ensemble4 = VotingClassifier(estimators4, voting='hard')
ensemble4.fit(X_train, y_train)
ensemble4.score(X_test, y_test)
```

```python
from mlens.ensemble import SequentialEnsemble
ensemble5 = SequentialEnsemble()
ensemble5.add('blend', [lgb.LGBMClassifier(), LogisticRegression(penalty='l1')])
ensemble5.add('stack', [lgb.LGBMClassifier(), lgb.LGBMClassifier(), LogisticRegression(penalty='l1')])
ensemble5.add_meta(LogisticRegression(penalty='l1'))
ensemble5.fit(X_train, y_train)
preds5 = ensemble5.predict(X_test)
print(accuracy_score(preds5, y_test))
```

*Let's try Tuning*

```python
from sklearn.model_selection import RandomizedSearchCV
def hypertuning_rscv(est, p_distr, nbr_iter,X_train, y_train):
    rdmsearch = RandomizedSearchCV(est, param_distributions=p_distr,n_jobs=-1, n_iter=nbr_iter, cv=9)
    rdmsearch.fit(X_train, y_train)
    ht_params = rdmsearch.best_params_
    ht_score = rdmsearch.best_score_
    return ht_params, ht_score

p_dist2 = {'learning_rate':[0.1, 0.2, 0.3, 0.5, 0.7, 0.9, 0.95],'num_leaves':[10, 15, 20, 25, 30, 35, 40, 47, 55, 60, 65, 70, 75, 80]}
lgb_parameters, lgb_ht_score = hypertuning_rscv(lgb.LGBMClassifier(), p_dist2, 40, X_train, y_train)
print(lgb_parameters, lgb_ht_score)
```

```python
clf9 = lgb.LGBMClassifier(num_leaves= '20',learning_rate= '0.3')
clf9.fit(X_train, y_train)
y_pred9 = clf9.predict(X_test)
print('Best Accuracy:', accuracy_score(y_pred9, y_test))
print('Best RMSE:', math.sqrt(mse(y_pred9, y_test)))
```
